````python
import yaml
import openxlab
from openxlab.dataset import info, query, get, download
````

key.yaml 格式

```yaml
# Access Key:
ak: <Access Key>

# Secret Key:
sk: <Secret Key>
```

````python
with open('key.yaml', 'r') as f:
    keys = yaml.safe_load(f.read())
ak = keys['ak']
sk = keys['sk']
````

# 进行登录，输入对应的AK/SK

````python
openxlab.login(ak=ak, sk=sk)
````

# 数据集信息查看

````python
info(dataset_repo='OpenDataLab/WanJuan1_dot_0')
````

# 数据集文件列表查看

````python
print(query(dataset_repo='OpenDataLab/WanJuan1_dot_0'))
````

# 数据集文件下载

````python
download(dataset_repo='OpenDataLab/WanJuan1_dot_0',source_path='/README.md', target_path='./README.md')
````
